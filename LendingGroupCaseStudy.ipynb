<div class="alert alert-block alert-info">
    <span style='font-family:Georgia'>
        <b>Observation: </b><br>Here date column is in string data type and we can not use standard pandas to_datetime method for converting these date columns into datetime format, Since only month and day ("May-16") is available. Need to create a custom logic for the same. 
    </span>    
</div>

```python
def convert_to_datetime(date):
    """
    Convert a given date string into a Python datetime object.

    Parameters:
    date (str): The input date string in one of the formats: '%y-%b' or '%b-%y'.

    Returns:
    datetime: The datetime object representing the parsed date.

    Raises:
    ValueError: If the input date string is not in a valid format.
    """
    try:
        
        datetime_object = datetime.strptime(date, '%y-%b')
    except ValueError:
        try:
            datetime_object = datetime.strptime(date, '%b-%y')
        except ValueError:
            raise ValueError("Invalid date format. Supported formats: '%y-%b' or '%b-%y'")
            
    return datetime_object
```

```python
loanDF.earliest_cr_line = loanDF.earliest_cr_line.apply(lambda x: convert_to_datetime(x))
loanDF.issue_d= loanDF.issue_d.apply(lambda x: convert_to_datetime(x))
loanDF.last_pymnt_d= loanDF.last_pymnt_d.apply(lambda x: convert_to_datetime(x))
loanDF.last_credit_pull_d= loanDF.last_credit_pull_d.apply(lambda x: convert_to_datetime(x))
```
