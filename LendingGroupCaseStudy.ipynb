<div class="alert alert-block alert-info">
    <span style='font-family:Georgia'>
        <b>Observation: </b><br>Here date column is in string data type and we can not use standard pandas to_datetime method for converting these date columns into datetime format, Since only month and day ("May-16") is available. Need to create a custom logic for the same. 
    </span>    
</div>

```python
def convert_to_datetime(date):
    """
    Convert a given date string into a Python datetime object.

    Parameters:
    date (str): The input date string in one of the formats: '%y-%b' or '%b-%y'.

    Returns:
    datetime: The datetime object representing the parsed date.

    Raises:
    ValueError: If the input date string is not in a valid format.
    """
    try:
        
        datetime_object = datetime.strptime(date, '%y-%b')
    except ValueError:
        try:
            datetime_object = datetime.strptime(date, '%b-%y')
        except ValueError:
            raise ValueError("Invalid date format. Supported formats: '%y-%b' or '%b-%y'")
            
    return datetime_object
```

```python
loanDF.earliest_cr_line = loanDF.earliest_cr_line.apply(lambda x: convert_to_datetime(x))
loanDF.issue_d= loanDF.issue_d.apply(lambda x: convert_to_datetime(x))
loanDF.last_pymnt_d= loanDF.last_pymnt_d.apply(lambda x: convert_to_datetime(x))
loanDF.last_credit_pull_d= loanDF.last_credit_pull_d.apply(lambda x: convert_to_datetime(x))
```

#### Annual income against Chargedoff_Proportion
- We will first create a new column for annual income category
- Then perform group by operation using annual income category and loan status and count the occurrences
- Then calculate the total loans and charged-off loans for each income category
- sort by the proportion of charged-off loans in descending order
- Finally plot annual income category and charged-off proportion using barplot.

```python
# Categorizing annual income into income ranges
loanDF['annual_inc_cats'] = pd.cut(loanDF['annual_inc'], [0, 20000, 40000, 60000, 80000, 1000000], labels=['0-20000', '20000-40000', '40000-60000', '60000-80000', '80000+'])
```

```python
loanDF[["annual_inc", "annual_inc_cats"]]
```

```python
# Grouping by income categories and loan status, counting occurrences, and reshaping data
inc_range_vs_loan = loanDF.groupby(['annual_inc_cats', 'loan_status'])['loan_status'].count().unstack().fillna(0).reset_index()
```

```python
# Calculating total loans and proportion of charged-off loans for each income category
inc_range_vs_loan['Total'] = inc_range_vs_loan['Charged Off'] + inc_range_vs_loan['Fully Paid']
inc_range_vs_loan['Chargedoff_Proportion'] = inc_range_vs_loan['Charged Off'] / inc_range_vs_loan['Total']
```

```python
# Sorting by the proportion of charged-off loans in descending order
inc_range_vs_loan.sort_values('Chargedoff_Proportion', ascending=False, inplace=True)
```

```python
inc_range_vs_loan
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='annual_inc_cats', y='Chargedoff_Proportion', data=inc_range_vs_loan)
plt.ylabel('Charged-off Proportion')
plt.xlabel('Annual Income')
plt.title('Proportion of Charged-off Loans by Annual Income')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

plt.show()
```

<div class="alert alert-block alert-info">
    <span style='font-family:Georgia'>
        <b>Insight: </b><br>Annual income category and charged-off proportion
        
- Income range 80000+ has less chances of charged off. Income range 0-20000 has high chances of charged off. Notice that with increase in annual income charged off proportion got decreased.

    </span>    
</div>
